```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
for course in COURSES:
    text = course["title"] + "." + course["desc"]
    course["vector"] = model.encode(text)
```

```python
TRANSCRIPT = [
    {"id": "CS135", "grade": 92},
    {"id": "CS136", "grade": 85},
    {"id": "MATH135", "grade": 78},
]
```

```python
user_vecs = []
weights = []
```

```python
for course in TRANSCRIPT:
    for courset in COURSES:
        if course["id"] == courset["id"]:
            user_vecs.append(courset["vector"])
            weights.append(course["grade"])
```

```python
import numpy as np
```

```python
user_vec = np.average(user_vecs, axis=0, weights=weights)
```

```python
def cosine_similarity(a, b):
    return float(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b)))
```

```python
results = []
```

```python
for course in COURSES:
    similarity = cosine_similarity(course["vector"], user_vec)
    results.append((course["id"], similarity))
```

```python
results.sort(key=lambda x: x[1], reverse=True)
```
